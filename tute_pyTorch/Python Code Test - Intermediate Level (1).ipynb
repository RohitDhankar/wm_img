#### 2.) Text Substitute
**Scenario:**

You are to write a function that accepts two arguments, a plain-text message and a number of letters to shift in the text. The function will return an encrypted string with all letters transformed and all punctuation and whitespace remaining unchanged.

For example, assume the input plain text is the following:
abcd xyz

If the shift value, n, is 4, then the encrypted text would be the following:
efgh bcd

##### HINT: use approproate function in the below code

```python
import string

def caesar(plain_text, shift_num=1):
   
    print("-shift_num--",shift_num)
    letters = string.ascii_lowercase
    print("letters--",letters)
    mask = letters[shift_num:] + letters[:shift_num]
    print("mask--",mask)
    
    trantab = str.maketrans(letters, mask)
    print("trantab--",trantab)
    
    return plain_text.translate(trantab)

#caesar('this is string', 4)
caesar('abc', 2)
```

#### 3.) Fix the below code

##### HINT: Expected output
1
2
3

```python
class ArrayList:
    def __init__(self, number_list):
        self.numbers = number_list

    def __iter__(self):
        self.pos = 0
        return self

    def __next__(self):
        if(self.pos < len(self.numbers)):
            self.pos += 1
            return self.numbers[self.pos - 1]
        else:
            raise StopIteration

array_obj = ArrayList([1, 2, 3])

it = iter(array_obj)

print(next(it)) #output: 1
print(next(it)) #output: 2
print(next(it)) #output: 3
```

#### 4.) Concatenate Strings
**Scenario:**

Write a function which takes a list of strings as an argument and returns, as a single-value result, the concatenation of all of these strings.

##### HINT: Fix error in the below code, Expected output: 'string-1_string-2_'

```python
def concat_function(args):
    temp_string = ''
    for word in args:
        temp_string += word
    return temp_string

concat_function(['string-1', 'string-2'])
```

#### 10.) Check below string is palindrome or not
##### HINTL: Debug the below code, Expected output: True

```python
def isPalindromeText(string):
     left,right=0,len(string)-1
     while right>=left:
             if string[left]==string[right]:
                      return False
             left+=1
             right-=1
             return True

isPalindromeText('MALAYALAM')
```
